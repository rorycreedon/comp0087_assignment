```python
# Download ECHR datasets
from download_hf_data import download_echr

download_echr('anon')
download_echr('non-anon')
```

```python
# Load train datasets into memory
import pandas as pd
train_df = pd.read_pickle('data/echr/anon_train.pkl')
train_df.head()
```

```python
from summarizer import Summarizer
model=Summarizer()
```

```python
from tqdm import tqdm

for split in ['valid', 'train']:
    for name in ['anon', 'non-anon']:
        print(name, split)
        df = pd.read_pickle(f'data/echr/{name}_{split}.pkl')
        df['summary'] = ''
        for i in tqdm(range(len(df))):
            df.loc[i, 'summary'] = model(df.loc[i, 'text'])
        df.to_pickle(f'data/echr/{name}_{split}_bert_summary.pkl')
```
